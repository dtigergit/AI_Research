```python
from openai import OpenAI
import os

# Set your token securely
DATABRICKS_TOKEN = "YOUR_DATABRICKS_TOKEN_HERE" # os.getenv('DATABRICKS_TOKEN')  # Or paste directly for dev/testing

# Set your Databricks model serving endpoint base
BASE_URL = "https://YOUR_DATABRICKS_SERVING_ENDPOINT_HERE.cloud.databricks.com/serving-endpoints"  # Update accordingly

# Check available models at Serving menu at left AI/ML section
MODEL = "databricks-qwen3-next-80b-a3b-instruct" # or "gpt-4o-mini", "dbrx-instruct", "databricks-llama-4-maverick", etc.

# Set up client
client = OpenAI(
    api_key=DATABRICKS_TOKEN,
    base_url=BASE_URL
)
```

```python
# Ask LLM model directly
response = client.chat.completions.create(
    model= MODEL,  
    messages=[
        {"role": "user", "content": "What is a database?"}
    ]
)

# Print the response
print(response.choices[0].message.content)
```

```python
# Put LLM request in a function
def ask_databricks_llm(prompt, model = MODEL):
    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "user", "content": prompt}
        ]
    )
    return response.choices[0].message.content

print(ask_databricks_llm("Explain the difference between RAG and fine-tuning."))
```
